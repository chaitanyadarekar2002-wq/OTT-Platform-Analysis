```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
netflix=pd.read_csv("netflix_titles.csv")
disney=pd.read_csv("disney_plus_shows.csv")
amazon=pd.read_csv("amazon prime movies.csv")
```

```python
netflix.head()
```

```python
disney.head()
```

```python
amazon.head()
```

```python
amazon['Year of Release']=amazon['Year of Release'].fillna(0).astype(int)
```

```python
amazon.head()
```

```python
netflix["platform"] = "Netflix"
amazon["platform"] = "Amazon Prime"
disney["platform"] = "Disney+"
```

```python
netflix.columns = netflix.columns.str.lower().str.strip()
amazon.columns = amazon.columns.str.lower().str.strip()
disney.columns = disney.columns.str.lower().str.strip()
```

```python
print(amazon.columns)
```

```python
print("Netflix Columns:")
print(netflix.columns)
```

```python
combined_data = pd.concat([netflix, amazon, disney], ignore_index=True)
```

```python
print(combined_data.info())
```

```python
print("\nShape of Dataset:")
print(combined_data.shape)
```

```python
print(combined_data.isnull().sum())
```

```python
combined_data.drop_duplicates(inplace=True)
```

```python
print("\nShape After Removing Duplicates:")
print(combined_data.shape)
```

```python
type_count = combined_data['type'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(
    type_count,
    labels=type_count.index,
    autopct='%1.1f%%'
)
plt.title("Movies vs TV Shows")
plt.show()
```

```python
import seaborn as sns
heatmap_data = pd.crosstab(combined_data['platform'], combined_data['type'])
plt.figure(figsize=(6,4))
sns.heatmap(
    heatmap_data,
    annot=True,
    fmt='d'
)
plt.title("Platform vs Content Type")
plt.show()
```

```python
top_rating = combined_data['rating'].value_counts().head(10)
plt.figure(figsize=(10,6))
top_rating.plot(kind='bar')

plt.title("Content Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
growth = (
    combined_data.groupby(['release_year', 'platform'])
    .size()
    .reset_index(name='count')
)
plt.figure(figsize=(12,6))
sns.lineplot(
    data=growth,
    x='release_year',
    y='count',
    hue='platform'
)

plt.title("Platform Content Growth Over Time")
plt.xlabel("Release Year")
plt.ylabel("Content Count")

plt.show()
```

```python
combined_data.to_csv("ott_cleaned_data.csv", index=False)
```
